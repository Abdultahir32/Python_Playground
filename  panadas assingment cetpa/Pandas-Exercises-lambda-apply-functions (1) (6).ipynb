### Step 5. Create a lambda function that captalize strings.

```python
# df['Mjob'] = df['Mjob'].str.capitalize()
```

```python
for i in df:
    if df[i].dtype=='object':
        df[i] = df[i].apply(lambda x : x.capitalize())
```

```python
df
```

### Step 6. Convert  both Mjob and Fjob in Uppercase

```python
for i in df:
    if i=='Mjob' or i=='Fjob':
        df[i] = df[i].apply(lambda a:a.upper())
```

### Step 8. Create a function called majority that return a boolean value to a new column called legal_drinker

```python
def majority(y):
    if y>18:
        return True
    else:
        return False
df['legal_drinker'] = df['age'].apply(majority)
```
