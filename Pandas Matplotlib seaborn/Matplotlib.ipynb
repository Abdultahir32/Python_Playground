# Box Plot

```python
# Box plot = it is used on detect the outliers in the data
```

```python
df1
```

```python
df1['tip'].mean()
```

```python
df1['tip'].median()
```

```python
plt.boxplot(df1['tip']);
```

```python
0quantile.......0percentile
1quantile.......25percentile
2quantile.......50percentile
3quantile.......75percentile
4quantile.......100percentile
```

```python
# upper_limit = q3+1.5 * (IQR)       1.5 hai standard deviation 
# lower_limit = q1-1.5 * (IQR)
# IQR = Q3 - q1
```

```python
df1['tip'].describe()
```

```python
q1 = np.percentile(df1['tip'],25)       #df['tip'].quantile(0.25)
q1
```

```python
q3 = df1['tip'].quantile(0.75)
q3
```

```python
IQR = q3 - q1
IQR
```

```python
u = q3+1.5 * (IQR)
u
```

```python
l = q1-1.5 * (IQR)
l
```

```python
df1[df1['tip']>u]    # for upper limit detection 
```

```python
df1[df1['tip']<l]  # for lower limit detection 
```

# Handling Outliers
- Trimming
- Capping (Winsorization)

```python
df1
```

```python
df1[df1['tip']<u]
```

```python
# capping
```

```python
df1['tips'] = np.where(df1['tip'] > u, u,
         np.where(df1['tip'] <l,l,
                  df1['tip']))
```
