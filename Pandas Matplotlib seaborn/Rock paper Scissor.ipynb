```python
import streamlit as st
import random

# Page configuration
st.set_page_config(page_title="Rock Paper Scissors", page_icon="ğŸ®")

# Title
st.title("ğŸ® Rock, Paper, Scissors Game")
st.markdown("Let's play a quick game of Rock, Paper, Scissors!")

# Choices
choices = ['rock', 'paper', 'scissor']
emojis = {'rock': 'ğŸª¨', 'paper': 'ğŸ“„', 'scissor': 'âœ‚ï¸'}

# Score tracking in session
if 'player_score' not in st.session_state:
    st.session_state.player_score = 0
if 'computer_score' not in st.session_state:
    st.session_state.computer_score = 0

# User input
player_choice = st.radio("Choose your move:", choices, horizontal=True)

# Play button
if st.button("Play"):
    computer_choice = random.choice(choices)

    st.markdown(f"**You chose:** {emojis[player_choice]} {player_choice.capitalize()}")
    st.markdown(f"**Computer chose:** {emojis[computer_choice]} {computer_choice.capitalize()}")

    if player_choice == computer_choice:
        st.info("ğŸ¤ It's a Tie!")
    elif (player_choice == 'rock' and computer_choice == 'scissor') or \
         (player_choice == 'paper' and computer_choice == 'rock') or \
         (player_choice == 'scissor' and computer_choice == 'paper'):
        st.success("ğŸ‰ You Win!")
        st.session_state.player_score += 1
    else:
        st.error("ğŸ’» Computer Wins!")
        st.session_state.computer_score += 1

# Scores
st.subheader("ğŸ“Š Scoreboard")
st.write(f"**You:** {st.session_state.player_score} | **Computer:** {st.session_state.computer_score}")

# Reset
if st.button("Reset Score"):
    st.session_state.player_score = 0
    st.session_state.computer_score = 0
    st.success("Scores have been reset!")

# Footer
st.markdown("---")
st.markdown("ğŸ‘¨â€ğŸ’» Made by [Analyst Tahir](https://github.com/Abdultahir32)")
```
