# Filtering and Sorting Data

### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

### Step 2. Import the dataset *chipotle.tsv*

```python
 df = pd.read_table('chipotle.tsv')
df
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = df
chipo
```

### Step 4. How many products cost more than $10.00? 

Use `str` attribute to remove the $ sign and convert the column to proper numeric type data before filtering.


```python
chipo['item_price'] = chipo['item_price'].str.replace('$','').astype(float)
chipo
```

```python
chipo[chipo['item_price']>10].shape[0]
```

### Step 5. What is the price of each item? 

print a data frame with only two columns `item_name` and that column which contain each `item_price`

```python
chipo['each_item_price'] = chipo['item_price']/chipo['quantity']
```

```python
chipo
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values(['item_name'])
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo[chipo['each_item_price']==chipo['each_item_price'].max()]
```
