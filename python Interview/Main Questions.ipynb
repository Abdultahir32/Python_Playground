# Create a Bankaccount to check balance deposite and withdraw balance  

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance   # public variable for simplicity

    def deposit(self, amount):
        self.balance += amount
        return f"{amount} deposited. New balance = {self.balance}"

    def withdraw(self, amount):
        if amount > self.balance:
            return "Insufficient balance"
        self.balance -= amount
        return f"{amount} withdrawn. Remaining balance = {self.balance}"

    def check_balance(self):
        return f"Current balance = {self.balance}"


# Testing
acc = BankAccount("Tahir", 1000)
print(acc.deposit(500))
print(acc.withdraw(300))
print(acc.check_balance())
print(acc.withdraw(5000))
```

```python
a = [1,2,3,4,5,9,8,7,6,5]
ar = 10
s = []
seen = set()
for num in a:
    target = ar - num
    if target in seen:
        s.append((target, num))
    seen.add(num)

print(s)
```
