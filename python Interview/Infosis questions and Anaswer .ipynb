```python
month = int(input("enter month number: (1,12)"))
if month <1 or month >12:
    print('invalid month')
elif month >=3 and month <=5:
    print('spring')
elif month in (6,7,8):
    print('summer')
elif month in (9,10,11):
    print('Autumn')
else:
    print('winter')
```
